```python
import os

PATH = "/home/cc/shubham/LSMQueryDrivenCompaction/development/rocksdb/examples/__working_branch/experiments-E256-B16-Y0.1"
```

```python
inserts = list()
range_queries = list()

with open(os.path.join(PATH, "workload.txt")) as ff:
    for line in ff.readlines():
        if line[0] == 'I':
            inserts.append(int(line.split(' ')[1]))
        elif line[0] == 'S':
            _, start, end = line.split(' ')
            range_queries.append((int(start), int(end)))
```

```python
inserts.sort()
```

```python
keys_in_range = list()

for rangeq in range_queries:
    num = 0
    for ins in inserts:
        if str(ins) > str(rangeq[0]) and str(ins) < str(rangeq[1]):
            num += 1
    keys_in_range.append(num)

print(keys_in_range)
```
